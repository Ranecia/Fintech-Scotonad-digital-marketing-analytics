```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

df = pd.read_csv('FinTech_Scotland_Marketing_Analytics.csv')
df.head()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=df, x='Platform', y='Engagement_Rate', ci=None)
plt.title('Average Engagement Rate by Platform')
plt.ylabel('Engagement Rate (%)')
plt.xlabel('Platform')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='UX_Version', y='Conversions')
plt.title('Conversion Performance by UX Version')
plt.show()
```

```python
top_campaigns = df.sort_values(by='Conversions', ascending=False).head(5)
top_campaigns[['Campaign_Name', 'Platform', 'Conversions']]
```
